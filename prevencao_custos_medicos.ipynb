### Normalização
Quando definido como True, o espaço de feição é transformado usando o método definido no parâmetro normalized_method.

A normalização é uma técnica frequentemente aplicada como parte da preparação de dados para aprendizado de máquina. O objetivo da normalização é redimensionar os valores das colunas numéricas no conjunto de dados sem distorcer as diferenças nos intervalos de valores ou perder informações. Existem vários métodos disponíveis para normalização, por padrão o PyCaret usa z-score.

Lembrando que z-score se trata da subtração do valor pela média, seguido pela divisão do resultado pelo desvio padrão. z = (x – u)/s

outras opções disponíveis são:

* **minmax**: dimensiona e traduz cada recurso individualmente de modo que fique no intervalo de 0 a 1.
*  **maxabs**: dimensiona e traduz cada recurso individualmente de modo que o valor absoluto máximo de cada recurso seja 1,0. Ele não desloca/centraliza os dados e, portanto, não destrói nenhuma dispersão.
*  **robust**: e traduz cada característica de acordo com o intervalo Interquartil. Quando o conjunto de dados contém valores discrepantes, o dimensionador robusto geralmente fornece melhores resultados.

### Combinação de níveis raros
Às vezes, um conjunto de dados pode ter um recurso categórico (ou vários recursos categóricos) que possui um número muito alto de níveis (ou seja, recursos de alta cardinalidade). Se tal recurso (ou recursos) for codificado em valores numéricos, a matriz resultante será uma matriz esparsa. Isso não apenas torna o experimento lento devido ao aumento múltiplo no número de recursos e, portanto, no tamanho do conjunto de dados, mas também introduz ruído no experimento. A matriz esparsa pode ser evitada combinando os níveis raros no recurso (ou recursos) com alta cardinalidade. Isso pode ser feito no PyCaret usando o parâmetro combine_rare_levels na configuração.

* **combine_rare_levels**: bool, default = False
Quando definido como True, todos os níveis em recursos categóricos abaixo do limite definido no parâmetro rare_level_threshold são combinados como um único nível. Deve haver pelo menos dois níveis abaixo do limite para que isso tenha efeito. rare_level_threshold representa a distribuição percentil da frequência de nível. Geralmente, essa técnica é aplicada para limitar uma matriz esparsa causada por um alto número de níveis em recursos categóricos.
* **rare_level_threshold**: float, padrão = 0,1
Distribuição percentual abaixo da qual as categorias raras são combinadas. Só entra em vigor quando combine_rare_levels estiver definido como True.

É válido ressaltar que, haja vista que selecionamos a função, o pycaret já retorna algumas métricas relacionadas ao modelo, pois já passou pela etapa de separação de dados e validação. Observe que, assim que se impõe a setup é retornado seus parâmetros. Em Fold Generator = KFold e Fold Number	= 10, ou seja, 10 rodadas com  distribuições distintas entre treino e teste nos dados, nos retorna como é separado essa validação, e antes já separa o conjunto de dados em treino e teste. 

K é o número de partes que se divide os dados, utilizando na primeira rodada, a primeira parte como validação e o resto como treino, na segunda, usa-se a segunda como validação e o resto como treino, na k-esima rodada usa-se o k-esimo como validação e o resto como treino, totalizando k rodadas, todo o conjunto de dados é utilizado como treino e teste, sufuciente para ter mais estabilidade nos resultados. A métrica final é a média das métricas em cada rodada.

```python
# tunagem = ENCONTRA HIPERPARAMETROS QUE MELHORAM RSMLE E APLICAM AO MODELO JÁ RETORNANDO MÉTRICAS

rf_tunado = pyreg.tune_model(rf, optimize = 'RMSLE') # RSMLE MELHOROU PARA 0.35
```

A tunagem se trata da avaliação de parâmetros de maneira sistemática. Encontra qual a melhor combinação de valores que deve passar aos algoritmos no momento do treinamento. Testa por meio de um método de validação selecionado (kfold) o desempenho de cada possível combinação de hiperparâmetros.

No pycaret, esses hiperparâmetros já são encontrados e já acoplados são no modelo, já retornando as métricas encontradas com as rodadas de validação após a aplicação da tunagem no modelo. Acima, tem a tunagem do Random Forest, e em seguida, foi aplicado novamente o KFold, observe como a métrica de interesse, que no caso é RMSLE, já diminuiu bastante.

```python
# Comparando os modelos implementados, ordenando pelo RMSLE, e selecionando apenas os 15 melhores

melhores = pyreg.compare_models(sort = 'RMSLE', n_select = 15)  # MELHOR MODELO GBR, COM RMSLE DE 0.40, APESAR DE PERDER PARA RF TUNADO
```

Em Machine Learning, o prático é usar vários modelos para o treinamento, selecionando aqueles que adquirem melhor peformance para a construção do modelo final. Acima se seleciona todos os modelos, aplica kfold em cada um, e é retornado em dataset, as métricas de cada.

Essa função treina e avalia o desempenho de todos os estimadores disponíveis na biblioteca de modelos usando validação cruzada. A saída dessa função é uma grade de pontuação com pontuações médias validadas cruzadas. As métricas avaliadas durante o CV podem ser acessadas usando a função get_metrics. Métricas personalizadas podem ser adicionadas ou removidas usando as funções add_metric e remove_metric.

O compare_models retorna apenas o modelo de melhor desempenho com base nos critérios definidos no parâmetro de classificação. É Precisão para experimentos de classificação e R2 para regressão. Você pode alterar a ordem de classificação passando o nome da métrica com base na qual deseja fazer a seleção do modelo.

### Ignorar baixa variação
Às vezes, um conjunto de dados pode ter um recurso categórico com vários níveis, onde a distribuição de tais níveis é distorcida e um nível pode dominar outros níveis. Isso significa que não há muita variação nas informações fornecidas por esse recurso. Para um modelo de ML, esse recurso pode não adicionar muitas informações e, portanto, pode ser ignorado para modelagem. Isso pode ser feito no PyCaret usando o parâmetro ignore_low_variance na configuração. Ambas as condições abaixo devem ser atendidas para que um recurso seja considerado um recurso de baixa variação.
Contagem de valores exclusivos em um recurso/tamanho de amostra < 10%
Contagem do valor mais comum / Contagem do segundo valor mais comum > 20 vezes.
* PARÂMETROS
* **ignore_low_variance**: bool, default = False
Quando definido como True, todos os recursos categóricos com variações estatisticamente insignificantes são removidos do conjunto de dados. A variância é calculada usando a razão de valores únicos para o número de amostras e a razão do valor mais comum para a frequência do segundo valor mais comum.

### Remover multicolinearidade
A multicolinearidade (também chamada de colinearidade) é um fenômeno no qual uma variável de recurso no conjunto de dados é altamente correlacionada linearmente com outra variável de recurso no mesmo conjunto de dados. A multicolinearidade aumenta a variância dos coeficientes, tornando-os instáveis ​​e ruidosos para modelos lineares. Uma dessas maneiras de lidar com a multicolinearidade é descartar um dos dois recursos que são altamente correlacionados entre si. Isso pode ser obtido no PyCaret usando o parâmetro remove_multicollinearity na configuração.
* PARÂMETROS
* **remove_multicolinearity**: bool, default = False
Quando definido como True, as variáveis ​​com intercorrelações superiores ao limite definido no parâmetro multicollinearity_threshold são eliminadas. Quando dois recursos são altamente correlacionados entre si, o recurso menos correlacionado com a variável de destino é descartado.
* **multicolinearity_threshold**: float, padrão = 0,9
Limiar usado para descartar os recursos correlacionados. Só entra em vigor quando remove_multicolinearity é definido como True.

```python
clas = setup(data = df,
                    target = 'smoker',
                    normalize = True, 
                    ignore_low_variance = True,
                    remove_multicollinearity = True, 
                    multicollinearity_threshold = 0.95,
                    n_jobs = -1)
```
