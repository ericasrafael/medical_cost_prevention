### NormalizaÃ§Ã£o
Quando definido como True, o espaÃ§o de feiÃ§Ã£o Ã© transformado usando o mÃ©todo definido no parÃ¢metro normalized_method.

A normalizaÃ§Ã£o Ã© uma tÃ©cnica frequentemente aplicada como parte da preparaÃ§Ã£o de dados para aprendizado de mÃ¡quina. O objetivo da normalizaÃ§Ã£o Ã© redimensionar os valores das colunas numÃ©ricas no conjunto de dados sem distorcer as diferenÃ§as nos intervalos de valores ou perder informaÃ§Ãµes. Existem vÃ¡rios mÃ©todos disponÃ­veis para normalizaÃ§Ã£o, por padrÃ£o o PyCaret usa z-score.

Lembrando que z-score se trata da subtraÃ§Ã£o do valor pela mÃ©dia, seguido pela divisÃ£o do resultado pelo desvio padrÃ£o. z = (x â€“ u)/s

outras opÃ§Ãµes disponÃ­veis sÃ£o:

* **minmax**: dimensiona e traduz cada recurso individualmente de modo que fique no intervalo de 0 a 1.
*  **maxabs**: dimensiona e traduz cada recurso individualmente de modo que o valor absoluto mÃ¡ximo de cada recurso seja 1,0. Ele nÃ£o desloca/centraliza os dados e, portanto, nÃ£o destrÃ³i nenhuma dispersÃ£o.
*  **robust**: e traduz cada caracterÃ­stica de acordo com o intervalo Interquartil. Quando o conjunto de dados contÃ©m valores discrepantes, o dimensionador robusto geralmente fornece melhores resultados.

### CombinaÃ§Ã£o de nÃ­veis raros
Ã€s vezes, um conjunto de dados pode ter um recurso categÃ³rico (ou vÃ¡rios recursos categÃ³ricos) que possui um nÃºmero muito alto de nÃ­veis (ou seja, recursos de alta cardinalidade). Se tal recurso (ou recursos) for codificado em valores numÃ©ricos, a matriz resultante serÃ¡ uma matriz esparsa. Isso nÃ£o apenas torna o experimento lento devido ao aumento mÃºltiplo no nÃºmero de recursos e, portanto, no tamanho do conjunto de dados, mas tambÃ©m introduz ruÃ­do no experimento. A matriz esparsa pode ser evitada combinando os nÃ­veis raros no recurso (ou recursos) com alta cardinalidade. Isso pode ser feito no PyCaret usando o parÃ¢metro combine_rare_levels na configuraÃ§Ã£o.

* **combine_rare_levels**: bool, default = False
Quando definido como True, todos os nÃ­veis em recursos categÃ³ricos abaixo do limite definido no parÃ¢metro rare_level_threshold sÃ£o combinados como um Ãºnico nÃ­vel. Deve haver pelo menos dois nÃ­veis abaixo do limite para que isso tenha efeito. rare_level_threshold representa a distribuiÃ§Ã£o percentil da frequÃªncia de nÃ­vel. Geralmente, essa tÃ©cnica Ã© aplicada para limitar uma matriz esparsa causada por um alto nÃºmero de nÃ­veis em recursos categÃ³ricos.
* **rare_level_threshold**: float, padrÃ£o = 0,1
DistribuiÃ§Ã£o percentual abaixo da qual as categorias raras sÃ£o combinadas. SÃ³ entra em vigor quando combine_rare_levels estiver definido como True.

```python
# pode-se selecionar um modelo do dataset por meio de :

rf = pyreg.create_model('rf') #RMSLE 0.43
```

Ã‰ vÃ¡lido ressaltar que, haja vista que selecionamos a funÃ§Ã£o, o pycaret jÃ¡ retorna algumas mÃ©tricas relacionadas ao modelo, pois jÃ¡ passou pela etapa de separaÃ§Ã£o de dados e validaÃ§Ã£o. Observe que, assim que se impÃµe a setup Ã© retornado seus parÃ¢metros. Em Fold Generator = KFold e Fold Number	= 10, ou seja, 10 rodadas com  distribuiÃ§Ãµes distintas entre treino e teste nos dados, nos retorna como Ã© separado essa validaÃ§Ã£o, e antes jÃ¡ separa o conjunto de dados em treino e teste. 

K Ã© o nÃºmero de partes que se divide os dados, utilizando na primeira rodada, a primeira parte como validaÃ§Ã£o e o resto como treino, na segunda, usa-se a segunda como validaÃ§Ã£o e o resto como treino, na k-esima rodada usa-se o k-esimo como validaÃ§Ã£o e o resto como treino, totalizando k rodadas, todo o conjunto de dados Ã© utilizado como treino e teste, sufuciente para ter mais estabilidade nos resultados. A mÃ©trica final Ã© a mÃ©dia das mÃ©tricas em cada rodada.

```python
# tunagem = ENCONTRA HIPERPARAMETROS QUE MELHORAM RSMLE E APLICAM AO MODELO JÃ RETORNANDO MÃ‰TRICAS

rf_tunado = pyreg.tune_model(rf, optimize = 'RMSLE') # RSMLE MELHOROU PARA 0.35
```

A tunagem se trata da avaliaÃ§Ã£o de parÃ¢metros de maneira sistemÃ¡tica. Encontra qual a melhor combinaÃ§Ã£o de valores que deve passar aos algoritmos no momento do treinamento. Testa por meio de um mÃ©todo de validaÃ§Ã£o selecionado (kfold) o desempenho de cada possÃ­vel combinaÃ§Ã£o de hiperparÃ¢metros.

No pycaret, esses hiperparÃ¢metros jÃ¡ sÃ£o encontrados e jÃ¡ acoplados sÃ£o no modelo, jÃ¡ retornando as mÃ©tricas encontradas com as rodadas de validaÃ§Ã£o apÃ³s a aplicaÃ§Ã£o da tunagem no modelo. Acima, tem a tunagem do Random Forest, e em seguida, foi aplicado novamente o KFold, observe como a mÃ©trica de interesse, que no caso Ã© RMSLE, jÃ¡ diminuiu bastante.

```python
# Comparando os modelos implementados, ordenando pelo RMSLE, e selecionando apenas os 15 melhores

melhores = pyreg.compare_models(sort = 'RMSLE', n_select = 15)  # MELHOR MODELO GBR, COM RMSLE DE 0.40, APESAR DE PERDER PARA RF TUNADO
```

Em Machine Learning, o prÃ¡tico Ã© usar vÃ¡rios modelos para o treinamento, selecionando aqueles que adquirem melhor peformance para a construÃ§Ã£o do modelo final. Acima se seleciona todos os modelos, aplica kfold em cada um, e Ã© retornado em dataset, as mÃ©tricas de cada.

Essa funÃ§Ã£o treina e avalia o desempenho de todos os estimadores disponÃ­veis na biblioteca de modelos usando validaÃ§Ã£o cruzada. A saÃ­da dessa funÃ§Ã£o Ã© uma grade de pontuaÃ§Ã£o com pontuaÃ§Ãµes mÃ©dias validadas cruzadas. As mÃ©tricas avaliadas durante o CV podem ser acessadas usando a funÃ§Ã£o get_metrics. MÃ©tricas personalizadas podem ser adicionadas ou removidas usando as funÃ§Ãµes add_metric e remove_metric.

O compare_models retorna apenas o modelo de melhor desempenho com base nos critÃ©rios definidos no parÃ¢metro de classificaÃ§Ã£o. Ã‰ PrecisÃ£o para experimentos de classificaÃ§Ã£o e R2 para regressÃ£o. VocÃª pode alterar a ordem de classificaÃ§Ã£o passando o nome da mÃ©trica com base na qual deseja fazer a seleÃ§Ã£o do modelo.

```python
# Deploy BÃ¡sico - se trata de um corpo de interaÃ§Ã£o bÃ¡sica, que deve ser incorporado para ser utilizado pelo usuÃ¡rio
import pandas as pd

def Custo(age, sex, bmi, children, smoker, region):
    dados0 = {'age': [age], 'sex': [sex], 'bmi': [bmi], 'children': [children], 'smoker': [smoker], 'region': [region]}
    dados = pd.DataFrame(dados0)

    modelo = pyreg.load_model('modelo-charges')

    pred = float(pyreg.predict_model(modelo, data = dados)['Label'].round(2))
    valor = round(1.8*pred, 2)
    
    print(' ')
    print('Idade: {}'.format(age))
    print('Sexo: {}'.format(sex))
    print('IMC: {}'.format(bmi))
    print('CrianÃ§as: {}'.format(children))
    print('Fumante: {}'.format(smoker))
    print('RegiÃ£o: {}'.format(region))
    print(' ')
    print('Custo do Seguro: ${}'.format(pred))
    print('Valor de Venda: ${}'.format(valor))
```

### Ignorar baixa variaÃ§Ã£o
Ã€s vezes, um conjunto de dados pode ter um recurso categÃ³rico com vÃ¡rios nÃ­veis, onde a distribuiÃ§Ã£o de tais nÃ­veis Ã© distorcida e um nÃ­vel pode dominar outros nÃ­veis. Isso significa que nÃ£o hÃ¡ muita variaÃ§Ã£o nas informaÃ§Ãµes fornecidas por esse recurso. Para um modelo de ML, esse recurso pode nÃ£o adicionar muitas informaÃ§Ãµes e, portanto, pode ser ignorado para modelagem. Isso pode ser feito no PyCaret usando o parÃ¢metro ignore_low_variance na configuraÃ§Ã£o. Ambas as condiÃ§Ãµes abaixo devem ser atendidas para que um recurso seja considerado um recurso de baixa variaÃ§Ã£o.
Contagem de valores exclusivos em um recurso/tamanho de amostra < 10%
Contagem do valor mais comum / Contagem do segundo valor mais comum > 20 vezes.
* PARÃ‚METROS
* **ignore_low_variance**: bool, default = False
Quando definido como True, todos os recursos categÃ³ricos com variaÃ§Ãµes estatisticamente insignificantes sÃ£o removidos do conjunto de dados. A variÃ¢ncia Ã© calculada usando a razÃ£o de valores Ãºnicos para o nÃºmero de amostras e a razÃ£o do valor mais comum para a frequÃªncia do segundo valor mais comum.

### Remover multicolinearidade
A multicolinearidade (tambÃ©m chamada de colinearidade) Ã© um fenÃ´meno no qual uma variÃ¡vel de recurso no conjunto de dados Ã© altamente correlacionada linearmente com outra variÃ¡vel de recurso no mesmo conjunto de dados. A multicolinearidade aumenta a variÃ¢ncia dos coeficientes, tornando-os instÃ¡veis â€‹â€‹e ruidosos para modelos lineares. Uma dessas maneiras de lidar com a multicolinearidade Ã© descartar um dos dois recursos que sÃ£o altamente correlacionados entre si. Isso pode ser obtido no PyCaret usando o parÃ¢metro remove_multicollinearity na configuraÃ§Ã£o.
* PARÃ‚METROS
* **remove_multicolinearity**: bool, default = False
Quando definido como True, as variÃ¡veis â€‹â€‹com intercorrelaÃ§Ãµes superiores ao limite definido no parÃ¢metro multicollinearity_threshold sÃ£o eliminadas. Quando dois recursos sÃ£o altamente correlacionados entre si, o recurso menos correlacionado com a variÃ¡vel de destino Ã© descartado.
* **multicolinearity_threshold**: float, padrÃ£o = 0,9
Limiar usado para descartar os recursos correlacionados. SÃ³ entra em vigor quando remove_multicolinearity Ã© definido como True.
